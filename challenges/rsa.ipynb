### RSA Starter 6

$ C = M^{e_0} \bmod N_0$ encrypt with public key $e_0$

$H(M) => S = H(M)^{d_1} \bmod N_1 $ sign and "encrypt" with private key $d_1$

at other end , decrypt the message with $m = C^{d_0} \bmod N_0$ 

now using first party's public key $e_1$ calculate $s = S^{e_1} \bmod N_1$ which results in H(M)

we check if $H(m)=s$

If we have a situation where 
$N_1 = p_1 * q$ 
$N_2 = p_2 * q$ 

Computing $gcd(N_1,N_2) = q$ we can then compute $p, \phi, d$ and decrypt 

```python
from math import gcd
import itertools

c = '249d72cd1d287b1a15a3881f2bff5788bc4bf62c789f2df44d88aae805b54c9a94b8944c0ba798f70062b66160fee312b98879f1dd5d17b33095feb3c5830d28'
from Crypto.Cipher import PKCS1_OAEP
from Crypto.PublicKey import RSA
from Crypto.Util.number import long_to_bytes, bytes_to_long
keys = []
ciphertexts = []
for i in range(1,50+1): 
    with open('data/ron-wrong-whit-right/'+str(i)+'.pem', 'r') as f:
        key = RSA.importKey(f.read())
        keys.append(key)
    with open('data/ron-wrong-whit-right/'+str(i)+'.ciphertext','r') as f: 
        ciphertext = bytes.fromhex(f.read())
        ciphertexts.append(ciphertext)

keys = list(itertools.combinations(keys,2))
ciphertexts = list(itertools.combinations(ciphertexts,2))

for i, k in enumerate(keys):
    pk1, pk2 = k 

    q = gcd(pk1.n, pk2.n)
    
    if q!=1: 
        p1 = pk1.n//q 
        p2 = pk2.n//q 

        c1, c2 = ciphertexts[i]

        phi1 = (p1 - 1) * (q - 1)
        phi2 = (p2 - 1) * (q - 1)

        d1 = pow(pk1.e, -1, phi1)
        d2 = pow(pk2.e, -1, phi2)

        k1 = RSA.construct((pk1.n, pk1.e, d1))
        k2 = RSA.construct((pk2.n, pk2.e, d2))

        
        print(PKCS1_OAEP.new(k1).decrypt(c1))
        print(PKCS1_OAEP.new(k2).decrypt(c2))
```

### Bespoke Padding

This is Franklin-Reiter attack described in https://crypto.stanford.edu/~dabo/pubs/papers/RSA-survey.pdf

Suppose $M_1, M_2 \in \mathbb{Z}^*_N$ are two distinct messages satisfying $M_1 = f(M_2) \bmod N$ for some known polynomial $f \in \mathbb{Z}_N[X]$

Give encryption of these messages $C_1,C_2$ we can recover $M_1, M_2$ 

$C_1 = M_1^e \bmod N = f(M_2) \bmod N$

$f(M_2) - C_1 = 0 \bmod N$ 

We know that $M_2$ is a root of polyonimal 

$g_1(x) = f(x)^e - C_1 \in \mathbb{Z}_N[x]$

Simliarly

$C_2 = M_2^e \bmod N$ 

$M_2^e - C_2 = 0 \bmod N$ 

$M_2$ is a root of polyonimal 

$g_2(x) = x^e - C_2 \in \mathbb{Z}_N[x]$

The linear factor $x-M_2$ divides both polynomials 

so we compute $gcd$ of $g_1$ and $g_2$ and $M_2$ is found

### Blinding Light

2.2 Blinding on https://crypto.stanford.edu/~dabo/papers/RSA-survey.pdf 

Let $<N,d>$ be Bob's private key and $<N,e>$ be his corresponding public key. Suppose an adversary wants Bob's signature on message $M \in \mathbb{Z}^*_N$. Bob refuses to sign this message. Adversary can try picking a random $r\in \mathbb{Z}^*_N$ and sets $M' = r^eM \bmod N$ and asks Bob to sign this message. Bob may provide signature $S'$ on this message but recall that $S' = (M')^d \bmod N$. An adversary can now simply compute $S = \frac{S'}{r} \bmod N$ and obtain Bob's signature $S$ on the original $M$. Indeed 

$S^e = \frac{(S')^e}{r^e} = \frac{(M')^{ed}}{r^e} \equiv \frac{M'}{r^e} = M \bmod N$ 

```python
import telnetlib 
import json
from Crypto.Util.number import bytes_to_long, long_to_bytes
import random 
import math 



HOST = 'socket.cryptohack.org' 
PORT = 13376
tn = telnetlib.Telnet(HOST, PORT)


def readline():
    return tn.read_until(b"\n")

def json_recv():
    line = readline()
    return json.loads(line.decode())

def json_send(hsh):
    request = json.dumps(hsh).encode()
    tn.write(request)

def get_r(N): 
    while True: 
        r = random.randint(1, N-1)
        if math.gcd(r,N) == 1: 
            return r 



readline()

data = {'option': 'get_pubkey'}
json_send(data) 
response = json_recv() 

N = int(response['N'],16)
e = int(response['e'],16)

ADMIN_TOKEN = b"admin=True"

r = get_r(N)

forgery = long_to_bytes((pow(r,e,N) * bytes_to_long(ADMIN_TOKEN)) %N).hex()

data = {'option': 'sign', 'msg': forgery}
json_send(data)
response = json_recv()

s_forgery = int(response['signature'],16) 

s = long_to_bytes(s_forgery * pow(r, -1, N) % N).hex() 

data = {'option': 'verify', 'msg': ADMIN_TOKEN.hex(), 'signature': s}
json_send(data)
response = json_recv()

print(response['response'])
```
